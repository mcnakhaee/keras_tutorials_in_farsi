### انتقال یادگیری (یادگیری انتقالی)

زمانی که می خواهیم یک مسئله را با استفاده از یادگیری ماشین و یادگیری عمیق حل کنیم ممکن است با شرایطی مواجه شویم که در آن داده های کافی برای آموزش مدل نداشته باشیم و در نتیجه مدل ما به اندازه کافی خوب نمی شود. علاوه بر این ممکن است منابع محاسباتی کافی مانند GPU برای به دست آوردن نتایج مد نظر را نیز در اختیار نداشتم باشیم. در این حالت انتقال یادگیری به کمک ما می آید تا این چالش ها را حل کنیم.
یکی از قوی ترین ایده های موجود در یادگیری عمیق این است که ما می توانیم دانشی که شبکه عصبی از حل یک مسئله یاد گرفته است را بگیریم و آن را بر روی مسأله دیگر به کار ببریم. مثلا از دانشی که الگوریتم A  ما برای دسته بندی 1000 کلاس موجود در دیتاست ایمیج نت یاد گرفته است برای دسته بندی تصاویر X-ray با استفاده از مدل B استفاده کنیم. 

بگذارید برای روشن این مسئله یک مثال بزنیم. فرض کنید که  شما و یک نفر انگلیسی زبان قصد یادگیری زبان عربی را دارید. از آن جا که الفبای زبان عربی و زبان فارسی با هم تقریبا یکی هستند و کلمات زیادی بین این دو زبان مشترک هستند شما می توانید از دانش خود در خصوص زبان فارسی برای یادگیری زبان عربی استفاده کنید.و لازم نیست از صفر برای یادگیری عربی شروع کنید. اما برای کسی که زبان مادری اش انگلیسی هست این ماجرا کمی فرق می کند.

```python
import tensorflow as tf
from tensorflow import keras
import seaborn as sns
import numpy as np
%matplotlib inline
```

پایه ای ترین و عمومی ترین راه انتقال یادگیری این است که لایه آهای متصل انتهایی شبکه را تغییر دهیم به شکل که خروجی این لایه ها متناسب با مسئله ما باشد. برای این کار لایه آخر به همراه وزن ها حذف می کنیم و وزن دهی را به صورت تصادفی تنها برای آن لایه انجام می دهیم و تعداد خروجی های مسئله را به جای مثلا 1000 کلاس به صورت 2 کلاس در می آوریم

مثلاً فرض کنید که ما می خواهیم  یک شبکه عصبی را  برای یک مسأله دسته بندی تصاویر آموزش  دهیم. در این حالت ابتدا از یک مدل از پیش آموزش داده شده که برای دسته بندی نوع متفاوتی از تصاویر آموزش داده شده است استفاده می کنیم. سپس وزن های لایه آخر را به صورت تصادفی مقدار دهی می‌کنیم. بعد از آن مجموعه جدید داده های ورودی X و y را به آن می دهیم تا برای مسئله جدیدی که مد نظر داریم آن را آموزش دهیم. اگر داده های ما کم باشند آنگاه یکی  دو لایه لایه آخر را معمولا آموزش می دهیم.

Take layers from a previously trained model.
Freeze them, so as to avoid destroying any of the information they contain during future training rounds.
Add some new, trainable layers on top of the frozen layers. They will learn to turn the old features into predictions on a new dataset.
Train the new layers on your dataset.
A last, optional step, is fine-tuning, which consists of unfreezing the entire model you obtained above (or part of it), and re-training it on the new data with a very low learning rate. This can potentially achieve meaningful improvements, by incrementally adapting the pretrained features to the new data.

لایه های کراس یک مشخصه ``trainable`` دارند که وقتی مقدار آن را برابر با `False` قرار دهیم وزن های قابل آموزش آن لایه دیگر به روزرسانی و آموزش داده نمی شوند. اصطلاحا به این کار freeze کردن وزن های یک لای گفته می شود.


Instantiate a base model and load pre-trained weights into it.

Freeze all layers in the base model by setting trainable = False.

Create a new model on top of the output of one (or several) layers from the base model.

Train your new model on your new dataset

Note that an alternative, more lightweight workflow could also be:

Instantiate a base model and load pre-trained weights into it.

Run your new dataset through it and record the output of one (or several) layers from the base model. This is called feature extraction.

Use that output as input data for a new, smaller model.

```python
img_gen_train = keras.preprocessing.image.ImageDataGenerator(rescale=1./255,validation_split=0.2)

train_data = img_gen_train.flow_from_directory('../data/flowers',
                          batch_size = 8,
                                  seed =42,
                                  target_size = (150,150),
                        subset = 'training' 
                                 )
val_data = img_gen_train.flow_from_directory('../data/flowers',
                          batch_size = 8,
                                  seed =42,
                                  target_size = (150,150),
                                subset = 'validation' )
```

لایه های آخر مدل از پیش آموزش داده شده احتمالا به کار نمی آیند چون حاوی ویژگی های سطح بالایی هستند که منحصر به داده های اصلی هستند.




```python
base_model = keras.applications.Xception( weights= 'imagenet',
                                         input_shape = (150,150,3),
                                         include_top = False)

```

```python
base_model.summary()
```

```python
def create_transfer_learning_model():
    base_model.trainable = False
    inputs = keras.layers.Input(shape = (150,150,3))
    x = base_model(inputs,training = False)
    x = keras.layers.Flatten()(x)
    x = keras.layers.Dense(128,activation='relu')(x)
    x = keras.layers.Dropout(0.2)(x)
    outputs = keras.layers.Dense(5,activation='softmax' )(x)
    final_model = keras.Model(inputs,outputs)
    return final_model
```

این که چه تعداد از لایه های آخر مدل از پیش آموزش داده شده را حذف کنیم بستگی به شباهت مسئله ای  که می خواهیم حل کنیم با مسئله اصلی  و حجم داده های ما دارد. هر چه داده های ما بیشتر باشند می توانیم لایه های بیشتری را اصطلاحا unfreeze کنیم.


```python

final_model.compile(optimizer = 'adam',
                   metrics=['accuracy'],
                   loss = 'categorical_crossentropy' )
```

```python
history = final_model.fit(train_data,
                          validation_data = val_data,                              
                        epochs=10)
```

```python
sns.lineplot(range(len(history.history['loss'])),history.history['loss'])
sns.lineplot(range(len(history.history['val_loss'])),history.history['val_loss'],color = 'indianred')
```

### Fine Tune کردن مدل ها
وقتی که وزن های لایه های جدیدی که به مدل از پیش آموزش داده شده اضافه کرده ایم به همگرایی رسید می توانیم یک کار دیگر انجام دهیم که عملکرد مدل حتی بهتر هم شود.

می توانیم به جای این که تنها لایه های اضافه شده را  آموزش دهیم وزن های کل شبکه  ( مدل از پیش آموزش داده شده) را مجددا با یک نرخ یادگیری کوچک آموزش دهیم و به روز رسانی کنیم یا به عبارتی دیگر لایه های مدل را unfreeze کنیم. 

به این کار اصطلاحا fine-tuning مدل می گوییم.



```python
base_model.trainable = True
```

```python
final_model.compile(optimizer = keras.optimizers.Adam(0.0001),
                   loss = 'categorical_crossentropy',
                   metrics=['accuracy'])
```

```python
history = final_model.(train_data,valid epochs=10)
```

```python
sns.lineplot(range(len(history.history['loss'])),history.history['loss'])
sns.lineplot(range(len(history.history['val_loss'])),history.history['val_loss'],color = 'indianred')
```

مسئله مهم این است که fine-tune کردن مدل را حتما بعد از مرحله انتقال یادگیری انجام دهیم و نه قبل از آن. چون که در غیر این صورت  وزن های مدل از پیش آموزش داده شده دیگر به درد ما نخواهند خورد. مسئله دوم این است که از یک نرح یادگیری کوچک برای به روزرسانی وزن های مدل استفاده کنیم چون که نمی خواهیم وزن های مدل تغییر زیادی کنند.
در نهایت باید حوسمان باشد که استفاده از fine-tuning ممکن است منجر به بیش برازش مدل شود.

### استفاده از callback ها برای کنترل فرآیند آموزش

برای ما مطلوب این است که بهترین مدل ( بالاترین صحت) را انتخاب کنیم و به بیش برازش نرسیم. یک کار خیلی ابتدایی برای دست یابی به این هدف (شاید) این است که مدل را مجددا و تنها برای 5 ایپاک آموزش دهیم. اما این کار عاقلانه نیست چون ممکن است وزن دهی اولیه تصادفی ما فرق کند و بهترین عملکرد به جای 5 ایپاک در 6 ایپاک به دست آید. برای همین بهتر است از یک رویکرد دیگر برای حل این مشکل استفاده کنیم که در کراس callback نامیده می شود. با callback ها می توانیم به کراس بگوییم که وضعیت هر مدل را در پایان هر ایپاک به خاطر بسپار. اگر وضعیت و عملکرد مدل در ایپاک (های) بعدی بهتر نشد فرآیند آموزش را متوقف کن و وضعیت بهترین ایپاک را به ما بازگردان

One or more shards that contain your model's weights.
An index file that indicates which weights are stored in a which shard.دیگر به کراس بگوییم که در پایان هر  ایپام وزن های آموزش داده شده را برای ما ذخیره کن

#### چه زمانی از یادگیری انتقالی استفاده کنیم؟

استفاده از انتقال یادگیری زمانی منظقی است که از مساله ای که می خواهیم  از آن یادگیری را انتقال دهیم داده های زیادی وجود داشته باشد و معمولا داده های نسبتا کمتری برای مساله ای که می خواهیم حل کنیم و روی آن انتقال دهیم وجود داشته باشد. مثلا 1میلیون عکس  از اشیا داریم و فقط 100 عکس از رادیولوژی داریم. به صورت برعکس اگر نسبت داده ها برعکس باشد استفاده از انتقال یادگیری منطقی نیست. چون هر مثال از عکس های رادیولوژی ارزش بیشتری نسبت به عکس اشیا دارد.


به علاوه،  پیش از این که از یک مدل از پیش آموزش داده شده  بر روی دیتاست جدید استفاده کنیم لازم است ببینیم که داده های اصلی که این مدل بر روی آن ها آموزش داده شده چه بوده‌اند. این کار را برای این آنجا می‌دهیم تا بررسی کنیم که  آیا این داده‌ها بر روی استفاده ما از این مدل از پیش آموزش داده شده  محدودیت‌هایی ایجاد می‌کنند یا خیر؟  یا آیا bias هایی وجود دارد یا خیر؟  مثلاً از آنجا مدل VGG 16 بر روی تصاویر مسابقه imagenet که اغلب دارای یک شی هستند آموزش داده شده است بهترین زمانی که می توان استفاده کرد وقتی است که تصاویری که می خواهیم بر روی وزن های از پیش آموزش داده  شده آن آموزش دهیم نیز دارای یک شی باشند.



یادگیری انتقالی زمان بهترین عملکرد را دارد که task ای که پیش رو داریم شباهت با مدل از پیش آموزش داده شده داشته باشد یا به عبارت فنی تر ویژگی های سطح پائین تر (لایه های اولیه شبکه) آن ها با هم با مشابه باشد.
